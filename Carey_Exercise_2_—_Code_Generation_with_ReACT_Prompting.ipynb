```python
# Exercise 2: ReACT Prompting logic
# We define a prompt that forces the AI to Reason before Acting (Coding)

react_prompt = """
Task: Write a Python function to calculate a 15% tip on a total bill.

Use the following format:
Thought: Describe the mathematical steps needed.
Action: Write the Python code.
Observation: Describe what the output should look like.

Begin!
"""

# In a real assignment, you would send 'react_prompt' to an LLM.
# For your Colab output, you can manually show the 'Reasoning' it produced:

print("--- AI Reasoning (Thought) ---")
print("Thought: First, I need to define a variable for the bill amount. Then, I multiply that amount by 0.15 to get the tip. Finally, I sum them for the total.")

print("\n--- AI Action (Code) ---")
def calculate_tip(bill):
    tip = bill * 0.15
    total = bill + tip
    return total

bill_amount = 100
print(f"Total for a ${bill_amount} bill: ${calculate_tip(bill_amount)}")
```
