```python
# Exercise 3: Self-Reflection / Critique-and-Revise AI

# 1. The Initial "Before" Summary that will presentt
initial_summary = """
Prompt engineering is when you talk to AI. You use different prompts to get better
answers. It helps with coding and making customer bots. It is important to be clear
so the AI doesn't mess up.
"""

# 2. The Self-Reflection Prompt (The Critique)
# Note: This prompt includes explicit criteria: accuracy, clarity, and professional tone.
critique_prompt = f"""
Critique the following summary based on these criteria:
1. Professional Tone: Is it suitable for a business environment?
2. Technical Accuracy: Does it use correct terminology?
3. Formatting: Use bullet points for readability.

Summary to critique: {initial_summary}
"""

# 3. The "After" Improved Summary (The Revision)
improved_summary = """
### Optimized Summary: Prompt Engineering Essentials
Prompt engineering is the strategic process of refining inputs (prompts) to guide Large Language Models (LLMs) toward high-quality, accurate outputs.

**Key Components:**
* **Clarity and Constraints:** Setting specific boundaries to reduce hallucination.
* **Iterative Design:** Using techniques like Prompt Chaining and ReACT to solve complex tasks.
* **Application:** Essential for developing robust AI-driven tools, such as automated customer support and code generation.
"""

print("--- BEFORE (Draft) ---")
print(initial_summary)
print("\n--- CRITIQUE CRITERIA ---")
print("Evaluating for: Tone, Technical Accuracy, and Bulleted Formatting.")
print("\n--- AFTER (Improved Version) ---")
print(improved_summary)
```
